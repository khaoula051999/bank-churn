<b>Description des Variables du Dataset </b>

Nous pouvons observer que le dataset contient des variables catégorielles et continues. De plus, il n'y a pas de valeurs nulles.

- **RowNumber** : Correspond au numéro d'enregistrement (ligne) et n'a pas d'effet sur le résultat.
- **CustomerId** : Contient des valeurs aléatoires et n'a pas d'effet sur le départ du client.
- **Surname** : Le nom de famille d'un client n'a pas d'impact sur sa décision de quitter la banque.
- **CreditScore** : Peut influencer le churn des clients, car un client avec un score de crédit élevé est moins susceptible de quitter la banque.
- **Geography** : La localisation du client peut affecter sa décision de quitter la banque.
- **Gender** : Il est intéressant d'explorer si le genre joue un rôle dans le départ d'un client.
- **Age** : Certainement pertinent, car les clients plus âgés sont moins susceptibles de quitter leur banque que les plus jeunes.
- **Tenure** : Fait référence au nombre d'années pendant lesquelles le client est client de la banque. Normalement, les clients plus anciens sont plus loyaux et moins susceptibles de quitter une banque.
- **Balance** : Un bon indicateur du churn des clients, car les personnes avec un solde plus élevé sont moins susceptibles de quitter la banque par rapport à celles avec des soldes plus bas.
- **NumOfProducts** : Fait référence au nombre de produits qu'un client a achetés via la banque.
- **HasCrCard** : Indique si un client possède ou non une carte de crédit. Cette colonne est également pertinente, car les personnes avec une carte de crédit sont moins susceptibles de quitter la banque.
- **IsActiveMember** : Les clients actifs sont moins susceptibles de quitter la banque.
- **EstimatedSalary** : Comme pour le solde, les personnes avec des salaires plus bas sont plus susceptibles de quitter la banque par rapport à celles avec des salaires plus élevés.
- **Exited** : Indique si le client a quitté la banque ou non.
- **Complain** : Indique si le client a fait une plainte ou non.
- **Satisfaction Score** : Score fourni par le client pour la résolution de sa plainte.
- **Card Type** : Type de carte détenue par le client.
- **Points Earned** : Points gagnés par le client pour l'utilisation de la carte de crédit.

La distribution des variable

```python
# Visualisation de la distribution des variables numériques
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
# Visualisation de la distribution des variables numériques
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numeric_features):
    plt.subplot(5, 3, i + 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution de {feature}')
plt.tight_layout()
plt.show()
#plt.savefig("/content/drive/MyDrive/variableNum.png")
```

```python
# Pairplot
features_of_interest = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']
sns.pairplot(df[features_of_interest], hue='Exited')
plt.show()
#pairplot.savefig("/content/drive/MyDrive/pairplot.png")
```

```python
#variables catégorielles
features_of_interest = ['Geography', 'Gender', 'Tenure', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'Satisfaction Score', 'Card Type']

# Créer des histogrammes comparatifs pour chaque caractéristique
plt.figure(figsize=(15, 12))
for i, feature in enumerate(features_of_interest):
    plt.subplot(3, 3, i + 1)
    sns.countplot(data=df, x=feature, hue='Exited')
    plt.title(f'Variable {feature}')
plt.tight_layout()
plt.suptitle('Analysis Of Variable Exited', y=1.02)
plt.show()
#plt.savefig("/content/drive/MyDrive/variableCategoriale.png")
```

```python
#variable continue
continuous_features = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary', 'Point Earned']

# Créer des box plots comparatifs pour chaque caractéristique continue
plt.figure(figsize=(15, 12))
for i, feature in enumerate(continuous_features):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(data=df, x='Exited', y=feature)
    plt.title(f'Variable {feature}')
plt.tight_layout()
plt.suptitle('Analysis Of Variable Class', y=1.02)
plt.show()
```

```python
# Matrice de corrélation
# Sélectionner uniquement les colonnes numériques
numeric_df = df.select_dtypes(include=[np.number])

# Vérifier les colonnes numériques
print(numeric_df.columns)
# Matrice de corrélation pour les colonnes numériques
corr_matrix = numeric_df.corr()

# Visualisation de la matrice de corrélation
plt.figure(figsize=(10, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de Corrélation')
plt.show()

#plt.savefig("/content/drive/MyDrive/correlation_matrix.png")
```

###A Titre d'information pour interpréter une matrice de corrélation
* Valeur de Corrélation :
<br/>
	•	+1 : Corrélation positive parfaite. À mesure qu'une variable augmente, l'autre augmente proportionnellement.
  <br/>
•	0 : Aucune corrélation. Les variables ne montrent pas de relation linéaire.
 <br/>
•	-1 : Corrélation négative parfaite. À mesure qu'une variable augmente, l'autre diminue proportionnellement.
 <br/>
* Force de la Corrélation :
•	0.0 à 0.19 : Très faible corrélation
 <br/>
•	0.2 à 0.39 : Faible corrélation
 <br/>
•	0.4 à 0.59 : Corrélation modérée
 <br/>
•	0.6 à 0.79 : Corrélation forte
 <br/>
•	0.8 à 1.0 : Corrélation très forte



# Conclusions sur l'exploration des données



## Corrélations Notables :
1. Age et Exited : Corrélation de 0.29, indiquant que les clients plus âgés sont plus susceptibles de quitter.
2. Balance et Exited : Corrélation de 0.12, montrant une légère relation positive entre le solde des comptes et la probabilité de quitter.
3. IsActiveMember et Exited : Corrélation de -0.16, montrant que les membres actifs sont légèrement moins susceptibles de quitter.
4. Satisfaction Score et Exited : Corrélation de 0.01, suggérant une relation très faible entre la satisfaction et le départ des clients.

## Décisions Basées sur la Corrélation :

 ### Variables à Conserver :
 1. Age : Corrélation de 0.29 avec Exited, ce qui en fait une variable pertinente à conserver dans le modèle.
 2. IsActiveMember : Corrélation de -0.16 avec Exited, ce qui indique une importance pour prédire la rétention des clients actifs.
 3. Balance : Corrélation de 0.12 avec Exited, malgré une relation faible, elle peut être informative.

### Variables à Supprimer :
 1. Complain : Avec une corrélation de 1.0 avec Exited, elle doit être supprimée pour éviter un overfitting, car elle ne contient pas d'information supplémentaire par rapport à Exited.
 2. RowNumber : Ce numéro d'index n'est pas pertinent pour un modèle prédictif.
 3. CustomerId : Identifiant unique, inutile pour la prédiction.
 4. Point Earned : Variable non pertinente pour la prédiction, peut être supprimée.
